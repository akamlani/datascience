```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.tools.plotting import scatter_matrix
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold
from sklearn.cross_validation import train_test_split
from statsmodels.tools import eval_measures

from sklearn.metrics import roc_curve
from sklearn.metrics import auc

from pprint import pprint
import math

# read in csv data
df_train = pd.read_csv('data/train.csv')
#df_train = df_train[:5]
df_test = pd.read_csv('data/test.csv')
#df_test = df_test[:5]
df_train.head()
```

```python
# give a description of the data
def describe_data(df, data_des):
    features = df.columns
    n_samples, n_features = df.shape
    print "{0} Set Dimensions: {1}".format(data_des, df.shape)
    print "Features: {0}".format(df.columns)
    #print df.describe()
    #print df.info()
    return {"frame":df, "features": features, "sample_size": n_samples, 
            "feature_size": n_features, "description": data_des}

# normalize the data
# this will not include the target (survived)
# precondition: no strings, all converted to integers
def normalize_data(df):
    return (df - df.mean())/df.std()
    
# describe data
obj = describe_data(df_train, "Training Vector")
ext_obj = describe_data(df_test, "Test Vector")
#obj["frame"].head()
#ext_obj["frame"].head()
```

```python
# from the cabin field, extract the particular first letters
def extract_cabin(df):
    if pd.notnull(df.Cabin):
        cabins = df.Cabin.split(" ")
        for cabin in cabins:
            df['Cabin_' + str(cabin[0])] = int(1)
    else:
        df['Cabin_' + 'U'] = int(1)
    return df

# once cabin vectors have been set up, clean up the cabins
def clean_cabin(df, features):    
    if "Cabin_A" in features: 
        if pd.isnull(df.Cabin_A): df.Cabin_A = 0            
    if "Cabin_B" in features: 
        if pd.isnull(df.Cabin_B): df.Cabin_B = 0
    if "Cabin_C" in features: 
        if pd.isnull(df.Cabin_C): df.Cabin_C = 0
    if "Cabin_D" in features: 
        if pd.isnull(df.Cabin_D): df.Cabin_D = 0
    if "Cabin_E" in features: 
        if pd.isnull(df.Cabin_E): df.Cabin_E = 0
    if "Cabin_F" in features: 
        if pd.isnull(df.Cabin_F): df.Cabin_F = 0
    if "Cabin_G" in features: 
        if pd.isnull(df.Cabin_G): df.Cabin_G = 0
    if "Cabin_T" in features: 
        if pd.isnull(df.Cabin_T): df.Cabin_T = 0
    if "Cabin_U" in features: 
        if pd.isnull(df.Cabin_U): df.Cabin_U = 0
    return df
            
# clean and munge the data
def munge_data(df):
    # convert dummies
    dummies_emb = pd.get_dummies(df['Embarked'])
    df_ = pd.concat([df, dummies_emb.rename(columns=lambda x: 'Embarked_' + str(x))], axis=1)
    df_.drop('Embarked', axis=1, inplace=True)
    # single column for sex as female or not
    df_.rename(columns={'Sex':'Is Female'},inplace=True)
    df_['Is Female']=df_['Is Female']=='female'
    # Change PClass from 1,2,3 to features
    dummies_pclass = pd.get_dummies(df_['Pclass'])
    df_ = pd.concat([df_, dummies_pclass.rename(columns=lambda x: 'Pclass_' + str(x))], axis=1)
    df_.drop('Pclass', axis=1, inplace=True)
    # Ticket Feature does not seem to have a pattern: We will drop this before the model
    df_.drop('Ticket', axis=1, inplace=True)
    
    # Cabins
    # Create new feature vectors based on which cabins(s) individuals are in
    df_ = df_.apply(extract_cabin, axis=1) 
    df_ = df_.apply(lambda x: clean_cabin(x, df_.columns), axis=1)
    df_.drop('Cabin', axis=1, inplace=True)
    
    # Names
    # We will drop the Names for the Tttle
    
    # should we drop any 'na' rows: df = df.dropna()
    # should we ignore or imputate some missing values?            
    # TBD    

    # Fare
    # Should we add another feature based on fare: there are some that have a zero fare (children, staff, freeloaders)
    # Based on the Fare Cost we can associate it with the particular Passenger Class
    # TBD
            
    # Family (Alone=0, Family=1)
    # Should we create another feature combining: SibSp + Parch as binary feature if family is aboard
    # TBD

    # Age (Imputate)
    # Alternatively we could have taken the median or create a different heuristic
    # number of null Training Set: Age values = 891 - 714
    # number of null Test Set: Age values = 418 - 332

    #print "Describe Age Feature:\n{0}".format(df_.Age[df_.Age.notnull()].describe()) 
    avg_age = df_.Age[df_.Age.notnull()].mean()
    df_.Age.fillna(avg_age, inplace=True)
    #print "Describe Age Feature:\n{0}".format(df_.Age[df_.Age.notnull()].describe()) 
    
    return df_
       
obj["frame_"] = munge_data(obj["frame"])
ext_obj["frame_"] = munge_data(ext_obj["frame"])
#obj["frame_"].head()
#ext_obj["frame_"].head()
```

```python
# determine unique titles
def extract_titles(df):
    names_indices = df.Name.split(",")
    title_name = names_indices[1].lstrip().split(".")
    df["Title"] = title_name[0].lower() + "."
    return df

def create_feature_from_name(df):
    df['Is Mrs. Miss.'] = (df.Title == 'mrs.') | (df.Title == 'miss.') 
    return df

def correlate_titles(df):
    df = df.apply(extract_titles, axis=1) 
    titles = np.unique(df["Title"])
    dummies_titles = pd.get_dummies(df["Title"])
    corr_titles_survived = pd.concat([df["Survived"], dummies_titles], axis=1).corr()
    print "Correlation w/Survived:\n{0}\n".format(corr_titles_survived.Survived)
    return df

# Determine correlation based on training set
obj["frame_"] = correlate_titles(obj["frame_"])    

# Via Inspection Results: 
# {Mrs., Miss.} has a high correlation with 'Survived'
# {Capt.} has a negative correlation with 'Survived', so he mostly does go down with the ship
# Drop the 'Name' in favor of them (Mrs., Miss as new features)        
obj["frame_"] = obj["frame_"].apply(create_feature_from_name, axis=1)
obj["frame_"].drop(['Name', 'Title'], axis=1, inplace=True)

# As we don't have 'Survived' field in test set, extract data and create feature from analysis of Training set
ext_obj["frame_"] = ext_obj["frame_"].apply(extract_titles, axis=1)    
ext_obj["frame_"] = ext_obj["frame_"].apply(create_feature_from_name, axis=1)
ext_obj["frame_"].drop(['Name', 'Title'], axis=1, inplace=True)

#obj["frame_"].head()
#ext_obj["frame_"].head()
```

### Create Model for Logistic Regression
* Separate into Input and Target Vectors as Numpy arrays
* Split into Train and Test Vector Set
* Perform Cross Validation via K-Folds
* Score the model
* Review the Performance of the Model
* Adapt the Model

```python
# Implement cross-validation for your classifer (logistic regression in this case)
# As 'Survived' field is boolean, it is a fit for a binary classifier
def cross_validate(X, y, classifier, k_fold) :
    # derive a set of (random) training and testing indices
    k_fold_indices = KFold( len(X), n_folds=k_fold, indices=True, shuffle=True, random_state=0)
    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :
        model = classifier(X[ train_slice  ],  y[ train_slice])
        k_score = model.score(X[ test_slice ], y[ test_slice ])
        k_score_total += k_score
    # return the average accuracy
    return (k_score_total/k_fold)        

# Determine the optimal number of k-folds 
# via inspection, we are not gaining much when k >> 50
def optimal_kfolds(X, y):
    cv_folds = []
    for k in range(2, 50):
        d = (k, cross_validate(X, y, LogisticRegression(C=1).fit, k) )
        cv_folds.append(d)

    df = pd.DataFrame(cv_folds, columns=["folds", "score"])    
    filters = df[df.score > 0.70]
    # Get maximum folds over range
    print "Maximum Fold Avg CV Accuracy:" 
    pprint(filters["score"].max())
    # sort as Top 5
    print "\nTop 5 Filtered Sorted Folds"
    sorted_df = filters.sort(['score'], ascending=[False])
    pprint(sorted_df[:5])
    #plot
    print "\nGraph of Folds vs Accuracy: "
    plt.xlabel('Folds')
    plt.ylabel('Avg Accuracy')
    plt.title('Avg Accuracy vs Folds', fontsize=14)
    plt.scatter(filters["folds"], filters["score"])
    plt.show()

    
# Optimal K-Folds Analysis
# The optimal folds based on our model seems to be at n_folds=4, with a cross validation score of 0.81
print "Non-Normalized Training Set"
obj["frame_v"] =  obj["frame_"].drop('Survived', axis=1, inplace=False)        
split_obj = optimal_kfolds(obj["frame_v"].values, obj["frame_"].Survived.values) 

# Does this change when we normalize the input
# Normalizing the Data didn't not chnage our result and even slightly lowered our cross validation score
print "Normalized Training Set"
obj["frame_v"] =  obj["frame_"].drop('Survived', axis=1, inplace=False)        
obj["frame_n"] =  normalize_data(obj["frame_v"])
split_obj_n = optimal_kfolds(obj["frame_n"].values, obj["frame_"].Survived.values)

# Does this model change when we change the regularization for overfitting
```

```python
# At this point we have a baseline for optimal n_folds in CV, Normalization Performance, and how are Model Scores
# Let us explore the individual feature performance to subset the model for better performance
# We have already reviewed the correlation via 'Pearsons', however correlation does not imply causation

print "Original Columns:\n{0}".format(obj["features"])
print "Current Feature Set:\n{0}".format(obj["frame_v"].columns)

def spin_model(df_v, df_target, plot=True):
    X = df_v.values
    y = df_target.values
    # Perform standard 80/20 split to evaluate the feature performance on the model
    # Alternatively we could have picked a particular k-fold split from our optimal k
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
    model_lr = LogisticRegression(C=1).fit(X_train, y_train)
    yhat = model_lr.predict(X_test)
    # Plot Model
    if plot:
        # Evaluate Feature Performance: Plot the coefficients
        feature_names_indices = np.arange(len(df_v.columns))
        plt.bar(feature_names_indices, model_lr.coef_.ravel())
        plt.title('Logistic Regression Feature Importance')
        plt.xlabel('Feature')
        plt.ylabel('Coefficient')    
        _ = plt.xticks(feature_names_indices + 0.5, df_v.columns, rotation='vertical')
    return model_lr, X_test, y_test, yhat
        
        
# Base Set
_ = spin_model(obj["frame_v"], obj["frame_"].Survived)
# Results from Base Set:
# Based on evaluation of the Feature Coefficients, it looks like the following features are dominating:
# {'Is Female', 'Is Mrs. Miss.', 'Pclass_1', 'Cabin_E'}
# And this somewhat corresponds to our initial model when we look at correlation coefficent with 'survived'.
```

```python
# Adjust Model Base on Minimizing Features and reviewing cross validation score from Logistic Regression
print "Sub Features Modeling"
obj["sub_frame_v"] = obj["frame_v"][['Cabin_D', 'Cabin_E', 'Cabin_F', 'Is Female', 
                                     'Pclass_1', 'Pclass_2', 'Pclass_3', 'Is Mrs. Miss.']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values) 

obj["sub_frame_v"] = obj["frame_v"][['Cabin_E', 'Is Female', 
                                     'Pclass_1', 'Pclass_2', 'Is Mrs. Miss.']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values) 

obj["sub_frame_v"] = obj["frame_v"][['Cabin_A', 'Cabin_D', 'Cabin_E', 'Cabin_F', 
                                     'Embarked_C', 'Embarked_Q', 'Fare', 'Is Female', 
                                     'Pclass_1', 'Pclass_2', 'Is Mrs. Miss.']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values) 

obj["sub_frame_v"] = obj["frame_v"][['Cabin_D', 'Is Female', 'Pclass_1', 'Is Mrs. Miss.']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values)

# Now adjust based on 'Pearsons' Coefficient with minimized feature set from feature importance
# In order of correlation: {Is Female, Is Mrs. Miss., Pclass_1, Fare, Embarked_C, Cabin_B, Cabin_D, Cabin_E}
print "Sub Features Modeling based on Pearsons Coefficient"
obj["sub_frame_v"] = obj["frame_v"][['Is Female', 'Is Mrs. Miss.', 'Pclass_1', 'Fare', 'Embarked_C', 'Cabin_B']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values) 

obj["sub_frame_v"] = obj["frame_v"][['Is Female', 'Is Mrs. Miss.', 'Pclass_1']]
split_obj = optimal_kfolds(obj["sub_frame_v"].values, obj["frame_"].Survived.values) 

# Unfortunately, Reducing our model to a minimal number of features did not give us a better score
# We will keep our base model, it also keeps our complexity of optimal k down
```

```python
# Lets see how are model performs on data we have not trained on
# We don't have the expected response data, but we can see what the residuals are between y_test - yhat
# We have already munged the test set (test.csv) and while we have no target labels, lets see how close we are
# Revert back to the Base Model where we achieve our best Cross Validation Score

# not we are dropping Cabin_T as a feature to compare equivalent data for residuals in predictions w/test data (test.csv)
# we are also not splitting the test.csv as we have no target values to split before the model
# we could split the test.csv itself if wanted
obj["frame_v_"] =  obj["frame_v"].drop('Cabin_T', axis=1, inplace=False)       
model_lr, X_test, y_test, yhat = spin_model(obj["frame_v_"], obj["frame_"].Survived, 0)
MSE = eval_measures.mse(yhat, y_test, axis=0)    
RMSE = eval_measures.rmse(yhat, y_test, axis=0)
print "{0} Model Score: {1}: RMSE: {2}, MSE: {3}".format(obj["description"], model_lr.score(X_test, y_test), RMSE, MSE)    
# See how the model predicts when we feed in the test.csv munged data
# If would be better to take a random distirubtion of sample comparision
yhat_ = model_lr.predict(ext_obj["frame_"].values)
MSE = eval_measures.mse(yhat_[:len(y_test)], y_test, axis=0)    
RMSE = eval_measures.rmse(yhat_[:len(y_test)], y_test, axis=0)
print "{0} : RMSE: {1}, MSE: {2}".format(ext_obj["description"], RMSE, MSE)   
# Take this predicted data and save to csv file (Passenger ID, Survived) for Kaggle
results = np.column_stack((ext_obj["frame_"]["PassengerId"], yhat_))
df_results = pd.DataFrame(results.astype('int'),columns=['PassengerID','Survived'])
df_results.to_csv('titanic_results.csv',index=False)
```

### ROC Curve
* Vary the threshold for survival
* Plot of ROC Curve
* AUC for model

```python
# Create an ROC curve for your logistic regression by varying the threshold value for survival

# convert the test split from object type to float type
X_test_ = X_test.astype(float)
# Probability Estimates of all classes ordered by label of classes [n_samples, n_classes]
target_predicted_proba = model_lr.predict_proba(X_test_)

# Compute Receiver operating characteristic (ROC) 
# y_test as test slice of 'Survived' as True vs predicted 
fpr, tpr, thresholds = roc_curve(y_test, target_predicted_proba[:, 1])
# Compute Area Under the Curve (AUC)
roc_auc = auc(fpr, tpr)
print "Area Under Curve: {0}".format(roc_auc)

# Plot the ROC Curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
```

```python
# By default the model prediction function uses a threshold of 50%
# Lets find a disciribution of where are prediction probabilities are
prob_dist = np.arange(0,1,0.1)
labels = ["Thresh_" + str(proba_i) for proba_i in prob_dist]

def prob_thresh(df):
    state = df.prob > df.thresh
    label_idx = math.trunc(df.thresh * 10)
    df["Thresh_" + str(prob_dist[label_idx])] = state
    return df

df_prob = pd.DataFrame(zip(target_predicted_proba[:, 1], thresholds), columns=["prob", "thresh"])
df_prob = df_prob.apply(prob_thresh, axis =1)
df_prob = df_prob.fillna(0)
prob_thresh_counts = [(label, df_prob[label].value_counts()) for label in labels if label in df_prob]
pd.DataFrame(prob_thresh_counts, columns=['threshold', 'value counts'])     

# Result
# Unfortunately it doesn't look like we probability prediction was above the threshold very often when the
# threshold is > 0.5.  It may be worth increasing the threshold to 0.6 as we got the same results, but 
# afterwards the probability above being the threshold is not very good when you consider the number of 
# times that the probability was below the threshold.
```
